# **Pandas 2**

## **Add Remove Rows and Columns**

```python
import pandas as pd
```

```python
people = {
    "first": ["Ramesh", 'Suresh', 'Mohan', 'Rajesh'], 
    "last": ["Kapadiya", 'Paul', 'Kumar', 'Kumar'], 
    "email": ["RameshK@gmail.com", 'Paulsuresh@email.com', 'Kmohan@email.com', 'kkumarrajesh@gmail.com'],
    "job_role": ["Machine Learning Engineer", "Web Developer", "Data Scientist", 'Python Developer']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# Now if we want to add column name 'first' and 'last' and make it a new column called Full Name

df['Full Name'] = df['first'] + ' ' + df['last']
```

```python
df
```

```python
# Now we want to remove the two column  called 'first' and 'last'
# We will use drop function for this operation
df.drop(columns = ['first', 'last'], inplace=True)
```

```python
df
```

```python
# Now if we want to split the full name column into two column by first_name and last_name
# We can do this by using the expand method
df['Full Name'].str.split(' ',expand=True)
```

```python
# Now to assign the name of the columns
df[['first', 'last']] = df['Full Name'].str.split(' ', expand=True)
```

```python
df
```

```python
# Now to add values in column
# Leat say we want to add name in First Name column
# We will use append method for doing that.
df.append({'first':'Viresh'}, ignore_index=True)
```

```python
# We can also merge two dataframe together
# For that we need to create new dataframe
# Let's create a new dataframe

people2 = {
    "first": ["Tony", 'Steve'], 
    "last": ["Stark", 'Roger'], 
    "email": ["ironman@avengers.com", 'Cat@avenger.com']
}
```

```python
df2 = pd.DataFrame(people2)
```

```python
df2
```

```python
df.append(df2, ignore_index=True)
```

```python
df2.append(df, ignore_index=True)
```

```python
# To change it permanently we can assign like this
df = df.append(df2, ignore_index=True)
```

```python
# Removing rows
df.drop(index=4)
```

```python
# Now suppose if we want to drop all of the rows where last name is 'Kumar'

df.drop(index=df[df['last'] == 'Kumar'].index)
```

```python
# More conviently way to do 
filt = df['last'] == 'Kumar'
df.drop(index=df[filt].index)
```

## **Sorting Data**

```python
df
```

```python
# Sort this by last name

df.sort_values(by='last')
```

```python
# If we want to sort this in descending order 

df.sort_values(by='last', ascending=False)
```

```python
# Sort last name in desending order and first name in ascending order

df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```

```python
df
```
