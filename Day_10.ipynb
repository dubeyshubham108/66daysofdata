# **Seaborn**

Seaborn is also a data visualization library built on Matplotlib. 

## **Seaborn Scatter Plot**

```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading dataset of titanic

titanic_df = sns.load_dataset("titanic")
titanic_df
```

```python
# Draw scatter plot between age and fare

sns.scatterplot(x="age", y="fare", data=titanic_df)
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex")
plt.show()
```

```python
# To dividi our scatter plot into different category like man, woman and child 
# We can pass the style arguement with 'who' column

plt.figure(figsize=(10,9))
sns.scatterplot(x="age", y="fare", data=titanic_df, hue="sex", style="who")
plt.show()
```

```python
# To see how many person whether it is a male or a female were alive after the crash
# For that we can pass the alive column in hue and 

plt.figure(figsize=(16,9))
sns.scatterplot(x="who", y="fare", data=titanic_df, hue="alive", style="alive", size="who", sizes=(100,200))
plt.show()
```

## **Seaborn Distplot**

```python
# Loading the dataset

tips_df = sns.load_dataset('tips')
tips_df
```
