## **More Operations on Rows and Columns** 

For this operation first we will create a small dataframe and work on it. 

```python
# Creating a dataframe from dictionary

people = {
    "first": ["Ramesh", 'Suresh', 'Mohan', 'Rajesh'], 
    "last": ["Kapadiya", 'Paul', 'Kumar', 'Kumar'], 
    "email": ["RameshK@gmail.com", 'Paulsuresh@email.com', 'Kmohan@email.com', 'kkumarrajesh@gmail.com'],
    "job_role": ["Machine Learning Engineer", "Web Developer", "Data Scientist", 'Python Developer']
}
```

```python
# creating a dataframe
df = pd.DataFrame(people)
df
```

```python
# Now if we want to get the job_role column
people['job_role']
```

```python
# To get the first row 
# Remember that in iloc we are searching by integer locations
df.iloc[0]
```

```python
# Similarly if we want first and second row

df.iloc[[0, 1]]
```

```python
# if we want required column lets say we want email's first two column values

df.iloc[[0, 1], 2]
```

```python
# We can also use loc function
# loc function uses labels values

df.loc[[0, 1], 'email']
```

```python
# Similarly if we want last name and email we can pass a list of string

df.loc[[0, 1], ['last', 'email']]
```

## **Filtering Rows and Columns**



```python
# Lets say we want to filter out rows which have last name kumar
filt = df['last'] == 'Kumar'
```

```python
df[filt]
```

```python
#  We can also use AND OR operator by the symbol & |
# Let's say we want last name as Kumare and first name as Rajesh

filt = (df['last'] == 'Kumar') & (df['first'] == 'Rajesh')
```

```python
df.loc[filt]
```
